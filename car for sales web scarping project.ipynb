```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url='https://www.marutisuzukitruevalue.com/used-cars-in-thane'
```

```python
page= requests.get(url)
print(page)
```

```python
soup=BeautifulSoup(page.content,'html.parser')
#print(soup.prettify)
```

```python
result=soup.find('div',class_='row carBxGrid carLising2')
#print(result)
```

```python
lists=result.find_all('div',class_='fmpCardBox')
len(lists)
```

```python
car_data= []

for cars in lists:
    name=cars.find('div',class_='leftSec').a.get_text(strip=True)
    #print(name)
    location=cars.find('div',class_='location').get_text(strip=True).split('|')[0]
    #print(location)
    year_of_model=cars.find('div',class_='yearTxt').get_text(strip=True).split('|')[0]
    #print(year_of_model)
    fuel_type=cars.find('div',class_='yearTxt').get_text(strip=True).split('|')[1]
    #print(fuel_type)
    kilometer=cars.find('div',class_='yearTxt').get_text(strip=True).split('|')[2]
    #print(kilometer)
    price=cars.find('div',class_='priceTxt').get_text(strip=True).split('C')[0]
    #print(price)
    info=(name,location,year_of_model,fuel_type,kilometer,price)
    car_data.append(info)
    
```

```python
sale_car = pd.DataFrame(car_data, columns=['Name', 'Location', 'Year of Model', 'Fuel Type', 'Kilometer', 'Price'])
sale_car
```

```python
sale_car.to_excel('single_page_car.xlsx',index=False)
```

# pagination

```python


sale_car = pd.DataFrame(columns=['Name', 'Location', 'Year', 'Fuel Type', 'Kilometer', 'Price'])

for i in range(1, 6):
    url= f'https://www.marutisuzukitruevalue.com/used-cars-in-thane?page={i}'
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    result = soup.find('div', class_='row carBxGrid carLising2')
    lists = result.find_all('div', class_='fmpCardBox')
    
    for car in lists:
        name = car.find('div', class_='leftSec').a.get_text(strip=True)
        location = car.find('div', class_='location').get_text(strip=True).split('|')[0]
        year = car.find('div', class_='yearTxt').get_text(strip=True).split('|')[0]
        fuel_type = car.find('div', class_='yearTxt').get_text(strip=True).split('|')[1]
        kilometer = car.find('div', class_='yearTxt').get_text(strip=True).split('|')[2]
        price = car.find('div', class_='priceTxt').get_text(strip=True).split('C')[0]
        
        data = {'Name': name, 'Location': location, 'Year': year, 'Fuel Type': fuel_type,
                'Kilometer': kilometer, 'Price': price}
        sale_car= sale_car.append(data, ignore_index=True)

    
    
    
    
     


```

```python
print(sale_car)
```

```python
sale_car.to_excel('cars for sale.xlsx',index=False)
```
